```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = '/content/drive/MyDrive/보건의료데이터 공모전/merged_df.csv'
df = pd.read_csv(file_path)

# Select relevant columns
columns_needed = ['ID', 'age', 'BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']
df_mental = df[columns_needed]

# Filter for age 12–34 (based on international age)
df_mental = df_mental[df_mental['age'].between(12, 34)]

# Remove missing codes (8, 9) and NaNs
missing_codes = [8.0, 9.0]
condition = df_mental[['BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']].apply(
    lambda col: (~col.isin(missing_codes)) & col.notna()
)
df_cleaned = df_mental[condition.all(axis=1)]

# Define age group bins and labels (in English)
bins = [11, 15, 18, 34]
labels = ['Middle School', 'High School', 'Adult']
df_cleaned['age_group'] = pd.cut(df_cleaned['age'], bins=bins, labels=labels, right=True)

# Summary check
print(f"Rows after age filtering: {len(df_mental)}")
print(f"Rows after removing missing values: {len(df_cleaned)}")

# Mental health columns
mental_columns = ['BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']

# Plot and numeric summary
summary_tables = {}

for col in mental_columns:
    # Plot
    plt.figure(figsize=(7, 4))
    plot_data = df_cleaned.groupby('age_group')[col].value_counts(normalize=True).rename('Proportion').reset_index()
    sns.barplot(data=plot_data, x='age_group', y='Proportion', hue=col)
    plt.title(f'{col} - Distribution by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Proportion')
    plt.legend(title='Response')
    plt.tight_layout()
    plt.show()

    # Summary table
    summary = df_cleaned.groupby(['age_group', col]).size().reset_index(name='Count')
    total_by_group = df_cleaned.groupby('age_group').size().rename('Total').reset_index()
    merged = pd.merge(summary, total_by_group, on='age_group')
    merged['Proportion'] = (merged['Count'] / merged['Total']).round(3)

    # Pivot for display
    pivot = merged.pivot(index='age_group', columns=col, values='Proportion')
    summary_tables[col] = pivot

# Display summary tables
from IPython.display import display
for col, table in summary_tables.items():
    print(f"\n== {col} - Response Proportion by Age Group ==")
    display(table)
```

```python
# 전체 기준 BP1 응답 비율 표
bp1_counts = df_cleaned['BP1'].value_counts().sort_index()
bp1_percent = (bp1_counts / bp1_counts.sum() * 100).round(2)

# 하나의 DataFrame으로 정리
bp1_summary = pd.DataFrame({
    'Response': bp1_percent.index,
    'Count': bp1_counts.values,
    'Proportion (%)': bp1_percent.values
})

# 결과 출력
print("=== BP1 Overall Response Distribution ===")
print(bp1_summary)
```

```python
# 연령대별 BP1 응답 비율 테이블 (비율)
bp1_age_dist = (
    df_cleaned.groupby(['age_group', 'BP1']).size()
    .groupby(level=0)
    .apply(lambda x: x / x.sum())
    .unstack()
    .round(3)
)

print("=== BP1 Response Proportion by Age Group ===")
display(bp1_age_dist)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

mental_columns = ['BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']
results = []

for col in mental_columns:
    contingency = pd.crosstab(df_cleaned['age_group'], df_cleaned[col])
    chi2, p, dof, expected = chi2_contingency(contingency)

    if p < 0.001:
        sig = '***'
    elif p < 0.01:
        sig = '**'
    elif p < 0.05:
        sig = '*'
    else:
        sig = ''

    results.append({
        'Variable': col,
        'Chi-square': round(chi2, 2),
        'p-value': f"{p:.2e}",
        'Significance': sig
    })

# 결과 정리
df_chi2_results = pd.DataFrame(results)
display(df_chi2_results)
```

```python
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

mental_columns = ['BP1', 'BP5', 'BP6_10', 'BP6_2']

# Cramér's V 매트릭스 생성
cramer_matrix = pd.DataFrame(index=mental_columns, columns=mental_columns)

for col1 in mental_columns:
    for col2 in mental_columns:
        v = cramers_v(df_cleaned[col1], df_cleaned[col2])
        cramer_matrix.loc[col1, col2] = round(v, 3)

# 숫자형으로 변환
cramer_matrix = cramer_matrix.astype(float)

# 결과 출력
print("Cramér’s V Matrix (Strength of Association between Variables):")
display(cramer_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(cramer_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title("Cramér's V Correlation Between Mental Health Variables")
plt.show()


'''| V 값 범위       | 연관성 수준 |
| ------------ | ------ |
| 0.00 \~ 0.10 | 거의 없음  |
| 0.10 \~ 0.20 | 약한 연관  |
| 0.20 \~ 0.30 | 중간 연관  |
| 0.30 이상      | 강한 연관  |
'''
```

```python
# 라이브러리 불러오기
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# 정신건강 관련 변수만 추출
mental_vars = ['BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']
df_cluster = df_cleaned[mental_vars].copy()

# 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)

# 최적 k 찾기 (엘보우 & 실루엣 점수)
inertia = []
silhouette = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled, kmeans.labels_))

# 엘보우 & 실루엣 점수 시각화
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette, 'go-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')

plt.tight_layout()
plt.show()

# 최적 k 선택 후 클러스터링 수행 (예: k=3)
k_opt = 3
kmeans_final = KMeans(n_clusters=k_opt, random_state=42)
df_cleaned['cluster'] = kmeans_final.fit_predict(X_scaled)

# 군집별 정신건강 평균값 확인
cluster_summary = df_cleaned.groupby('cluster')[mental_vars].mean().round(2)
print("Cluster-wise Mental Health Profile:")
display(cluster_summary)

# 클러스터 분포 시각화
plt.figure(figsize=(6, 4))
sns.countplot(data=df_cleaned, x='cluster')
plt.title('Number of Respondents per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 정신건강 변수 선택
mental_vars = ['BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7']
X = df_cleaned[mental_vars]

# 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 클러스터링 수행 및 결과 시각화
for k in [5, 6, 9]:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    clusters = kmeans.fit_predict(X_scaled)
    df_cleaned[f'cluster_{k}'] = clusters

    # 클러스터별 정신건강 응답 분포 보기
    print(f'\n=== Clustering with K={k} ===')
    for var in mental_vars:
        print(f'\n▶ {var} distribution by cluster (K={k})')
        tab = pd.crosstab(df_cleaned[f'cluster_{k}'], df_cleaned[var], normalize='index').round(3)
        display(tab)
```

```python
from scipy.stats import chi2_contingency

for k in [5, 6, 9]:
    print(f"\n=== Chi-square test results for K={k} ===")
    for var in mental_vars:
        contingency = pd.crosstab(df_cleaned[f'cluster_{k}'], df_cleaned[var])
        chi2, p, dof, _ = chi2_contingency(contingency)
        print(f"{var}: χ² = {chi2:.2f}, p = {p:.4e} {'***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else ''}")
```
