```python
# 한글 컬럼으로 바꾸기
a={'N_KINDG1':'식품군 분류', 'NF_SUGAR':'당 섭취량', 'NF_EN':'에너지 섭취량',
'NF_INTK':'식품별 섭취량','NF_INTK3':'3차 식품코드별 섭취량',
'N_FCODE':'식품코드'}
# 각 변수에 직접 적용
df16 = df16.rename(columns=a)
df17 = df17.rename(columns=a)
df18 = df18.rename(columns=a)
df19 = df19.rename(columns=a)
df20 = df20.rename(columns=a)
df21 = df21.rename(columns=a)
df22 = df22.rename(columns=a)
df23 = df23.rename(columns=a)
```

```python
for df in [df16, df17, df18, df19, df20, df21, df22, df23]:
    df['ID'] = df['ID'].astype(str).str.replace(r"^b'|^b\"|\'$|\"$", '', regex=True)
    df['식품코드'] = df['식품코드'].astype(str).str.replace(r"^b'|^b\"|\'$|\"$", '', regex=True)
    df['식품군 분류'] = df['식품군 분류'].astype(str).str.replace(r"^b'|^b\"|\'$|\"$", '', regex=True)
```

```python
for df in [df16, df17, df18, df19, df20, df21, df22, df23]:
  df['식품코드'] = df['식품코드'].astype(float)
  df['식품군 분류'] = df['식품군 분류'].astype(float)
```

```python
# 음료 + 추가 식품군 필터링 (식품군 분류가 3, 8, 18, 21, 22인 경우)
target_codes = [3, 8, 18, 21, 22] # 당류, 과일류, 우유류, 음료류, 주류

df16 = df16[df16['식품군 분류'].isin(target_codes)]
df17 = df17[df17['식품군 분류'].isin(target_codes)]
df18 = df18[df18['식품군 분류'].isin(target_codes)]
df19 = df19[df19['식품군 분류'].isin(target_codes)]
df20 = df20[df20['식품군 분류'].isin(target_codes)]
df21 = df21[df21['식품군 분류'].isin(target_codes)]
df22 = df22[df22['식품군 분류'].isin(target_codes)]
df23 = df23[df23['식품군 분류'].isin(target_codes)]
```

```python
# 리스트로 묶어서 concat
df_all = pd.concat([
    df16, df17, df18,
    df19, df20, df21, df22, df23
], ignore_index=True)
df_all.head, df_all.tail()
```
