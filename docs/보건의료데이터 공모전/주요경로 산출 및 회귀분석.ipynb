```python
import glob
import os
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

# 1. 폴더 경로 지정 (구글드라이브나 로컬 환경에 맞게 수정)
folder_path = '/content/drive/MyDrive/보건의료데이터 공모전/건강영양조사 데이터'

# 2. 해당 폴더 안의 모든 CSV 파일 경로 불러오기
file_list = glob.glob(os.path.join(folder_path, '*.csv'))

# 3. 각 파일을 읽어서 리스트에 저장
df_list = []
for file in file_list:
    df = pd.read_csv(file)
    df_list.append(df)

# 4. 세로 방향으로 병합
merged_df = pd.concat(df_list, ignore_index=True)

# 6. 확인
print("파일 개수:", len(file_list))
print("병합된 행 개수:", merged_df.shape[0])

merged_df.to_csv('/content/drive/MyDrive/보건의료데이터 공모전/merged_df.csv', index=False)


merged_df
```

```python
# 1. 건강영양조사에서 필요한 칼럼만 추출 (ID + 총당섭취량만)
sugar_df = merged_df[['ID', '나이', '총 당 섭취량(g)', '총 에너지 섭취량(kcal)']]  # total_sugar_g → 실제 변수명으로 수정하세요

file_path = '/content/drive/MyDrive/보건의료데이터 공모전/24시 회상데이터/df_all_nofilter.csv'  # 파일명에 맞게 수정
df_all = pd.read_csv(file_path)
# 2. 식품섭취조사 데이터에 병합 (ID 기준, left join)
path_df = pd.merge(df_all, sugar_df, on='ID', how='left')

df_final = path_df[path_df['당 섭취량'] > 0]
df_final = df_final[df_final['총 당 섭취량(g)'] > 0]
df_final = df_final[df_final['총 에너지 섭취량(kcal)'] > 0]


# 3. 확인
df_final
```

```python
df_young = df_final[(df_final['나이'] >= 12) & (df_final['나이'] <= 34)]

df_young
```

```python
# 당섭취 열량 계산 (1g 당류 = 4kcal)
df_young['당섭취열량(kcal)'] = df_young['당 섭취량'] * 4

# 1. 당 섭취량(g) 기준 비율 (퍼센트)
df_young['당섭취비율_g(%)'] = (df_young['당 섭취량'] / df_young['총 당 섭취량(g)']) * 100

# 2. 당 섭취 열량(kcal) 기준 비율 (퍼센트)
df_young['당섭취비율_kcal(%)'] = (df_young['당섭취열량(kcal)'] / df_young['총 에너지 섭취량(kcal)']) * 100

# 3. 해당 식품에서 당이 차지하는 열량 비율 (%)
df_young['당 에너지 비율 WHO(%)'] = (df_young['당섭취열량(kcal)'] / df_young['에너지 섭취량']) * 100

df_young
```

```python
# 보정 대상 식품코드 리스트
target_sugar_codes = [3024.0, 3025.0, 3026.0, 3027.0, 3029.0, 50243.0, 50244.0, 50248.0, 50251.0, ]

# 먼저 '보정 여부'라는 새로운 컬럼 생성 (기본값: '아님')
df_young['보정 여부'] = 'X'

# 보정 대상 행 필터링
mask = df_young['식품코드'].isin(target_sugar_codes)

# 보정 적용
df_young.loc[mask, '당섭취열량(kcal)'] = df_young.loc[mask, '에너지 섭취량']
df_young.loc[mask, '당 에너지 비율 WHO(%)'] = 100.0
df_young.loc[mask, '보정 여부'] = '설탕 kcal 보정 O'
```

```python
import pandas as pd

# 데이터 불러오기
df_nm = pd.read_csv('/content/drive/MyDrive/보건의료데이터 공모전/식품+정신건강.csv')  # 경로는 환경에 맞게 수정

# 분류 기준으로 당 섭취량 합산
df_nm_sum = df_nm.groupby('분류')['당 섭취량'].sum().reset_index()

df_nm_sum = df_nm_sum.sort_values(by='당 섭취량', ascending=False).reset_index(drop=True)

df_nm_sum
```

```python
# 1. 각 분류별 총 당 섭취량 계산
df_nm_mean1 = df_nm.groupby('분류')['당 섭취량'].sum().reset_index()
df_nm_mean1.columns = ['분류', '총 당 섭취량']

# 2. 각 분류별 섭취자 수 계산 (해당 분류를 섭취한 ID 수)
df_nm_mean2 = df_nm.groupby('분류')['ID'].nunique().reset_index()
df_nm_mean2.columns = ['분류', '섭취자 수']

# 3. 병합 후 1인당 평균 섭취량 계산
df_nm_mean = pd.merge(df_nm_mean1, df_nm_mean2, on='분류')
df_nm_mean['1인당 평균 섭취량(g)'] = df_nm_mean['총 당 섭취량'] / df_nm_mean['섭취자 수']

# 4. 섭취량 기준 정렬
df_nm_mean = df_nm_mean.sort_values(by='1인당 평균 섭취량(g)', ascending=False).reset_index(drop=True)

# 결과 출력
df_nm_mean
```

```python
# 1. 총 당 섭취량 대비 비율 계산
df_nm_mean['총 당 섭취 비율'] = df_nm_mean['총 당 섭취량'] / df_nm_mean['총 당 섭취량'].sum()

# 2. 누적 기여율 계산 (내림차순 정렬 상태 유지 필요)
df_nm_mean['누적 기여율'] = df_nm_mean['총 당 섭취 비율'].cumsum()

# 3. 누적 기여율이 80% 이하인 식품군 추출
top_routes = df_nm_mean[df_nm_mean['누적 기여율'] <= 0.80]

# 4. 결과 출력
print(f"✅ 주요 당 섭취 경로로 분류된 식품군 수: {len(top_routes)}개")
display(top_routes[['분류', '1인당 평균 섭취량(g)', '총 당 섭취량', '섭취자 수', '누적 기여율']])
```

**공변량 추가**

```python
import pandas as pd
from google.colab import drive

# ──────────────── 0. Google Drive 연동 ────────────────
drive.mount('/content/drive')

# ──────────────── 1. 공변량 처리 ────────────────
# 공변량 데이터 경로 (파일명과 경로를 정확히 확인할 것!)
merged_df = pd.read_csv('/content/drive/MyDrive/보건의료데이터 공모전/merged_df.csv')

# 필요한 열만 추출
cols = ['ID', 'sex', 'age', 'ho_incm5', 'BD1_11', 'BE5_1', 'sm_presnt', 'HE_BMI' ]
df = merged_df[cols].copy()

# 나이 필터: 만 12~34세
df = df[(df['age'] >= 12) & (df['age'] <= 34)]

# 음주 빈도 처리
alcohol_map = {
    1: 0,     # 전혀 마시지 않는다
    2: 0.5,   # 월 1회 미만
    3: 1,     # 월 1회
    4: 3,     # 월 2~4회
    5: 10,    # 주 2~3회 → 약 10회/월
    6: 16     # 주 4회 이상 → 약 16회/월
}
df['BD1_11'] = df['BD1_11'].map(alcohol_map)
df.loc[df['age'] < 19, 'BD1_11'] = 0  # 미성년자 전부 '전혀 마시지 않음'
df = df[~((df['age'] >= 19) & (df['BD1_11'].isna()))]  # 성인 중 결측 제거

# 근력운동 일수 처리: 비해당/무응답 제거
df = df[~df['BE5_1'].isin([88, 99])]

df.loc[df['age'] < 19, 'sm_presnt'] = 0
adult = df['age'] >= 19
df = df[~(adult & df['sm_presnt'].isna())]

# 컬럼명 변경
df.rename(columns={
    'sex': '성별',
    'age': '나이',
    'ho_incm5': '가구소득분위',
    'BD1_11': '음주빈도',
    'BE5_1': '주간근력운동일수',
    'sm_presnt' : '흡연여부',
    'HE_BMI' : 'HE_BMI'
}, inplace=True)

# ID 중복 제거
df = df.drop_duplicates(subset='ID')
print("공변량 데이터 행 개수:", len(df))

# ──────────────── 2. 식품+정신건강 데이터 불러오기 ────────────────
mental_food_df = pd.read_csv('/content/drive/MyDrive/보건의료데이터 공모전/df00f.csv')
mental_food_df = mental_food_df.drop_duplicates(subset='ID')

# ──────────────── 3. 병합 ────────────────
final = pd.merge(mental_food_df, df, on='ID', how='inner')

final.to_csv('/content/drive/MyDrive/보건의료데이터 공모전/final.csv', index=False)

print("최종 병합된 데이터 행 개수:", len(final))
final
```

```python
# 나이 기준으로 필터링
teen_df = final[(final['age'] >= 10) & (final['age'] < 19)]
twenty_df = final[(final['age'] >= 19) & (final['age'] < 35)]

teen_df.to_csv('/content/drive/MyDrive/보건의료데이터 공모전/teen.csv', index=False)
twenty_df.to_csv('/content/drive/MyDrive/보건의료데이터 공모전/young.csv', index=False)
```

```python
import statsmodels.formula.api as smf

# 칼럼명 정리
teen_df = teen_df.rename(columns={'주요경로_비율(%)': 'sugar_ratio'})
twenty_df = twenty_df.rename(columns={'주요경로_비율(%)': 'sugar_ratio'})

# ✅ 1. 10대 회귀 분석 (음주, 흡연 제외)
formula_teen = 'mental_composite ~ sugar_ratio + age + C(가구소득분위) + HE_BMI + 주간근력운동일수'
print("🔹 [10대 회귀분석 결과]")
model_teen = smf.ols(formula=formula_teen, data=teen_df).fit()
print(model_teen.summary())

# ✅ 2. 20세~34세 회귀 분석 (음주, 흡연 포함)
formula_twenty = 'mental_composite ~ sugar_ratio + age + C(가구소득분위) + 음주빈도 + 흡연여부 + HE_BMI + 주간근력운동일수'
print("🔹 [20세~34세 회귀분석 결과]")
model_twenty = smf.ols(formula=formula_twenty, data=twenty_df).fit()
print(model_twenty.summary())

```
