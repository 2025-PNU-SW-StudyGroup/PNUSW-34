```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. 응답값 그룹핑 함수
def group_response(series):
    return series.map({
        1: 'Disagree',
        2: 'Disagree',
        3: 'Neutral',
        4: 'Agree',
        5: 'Agree'
    }).value_counts().reindex(['Disagree', 'Neutral', 'Agree'], fill_value=0)

# 2. 항목별 그룹화된 응답 수 계산
env_living_grouped = group_response(dm5_1['q8_5'])
env_safety_grouped = group_response(dm5_1['q8_8'])
env_air_grouped = group_response(dm5_1['q8_9'])

# 3. x축 위치 지정
x = np.arange(3)  # Disagree, Neutral, Agree
width = 0.25

# 4. 그래프 그리기
fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(x - width, env_living_grouped.values, width, label='Living Environment', color='skyblue')
ax.bar(x, env_safety_grouped.values, width, label='Safety Services', color='orange')
ax.bar(x + width, env_air_grouped.values, width, label='Environmental Quality', color='lightgreen')

# 5. 라벨/범례 설정
ax.set_ylabel('Number of Respondents')
ax.set_title('Satisfaction Level by Category (Grouped)')
ax.set_xticks(x)
ax.set_xticklabels(['Disagree', 'Neutral', 'Agree'])
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 1. 필터링: 1인 가구 & 19~35세
dm5_1 = df_24_3[(df_24_3['DM5'] == 1) & (df_24_3['DM2'].between(19, 35))]

# 2. 구 매핑
gu_mapping = {
    1: "① 장안구",
    2: "② 권선구",
    3: "③ 팔달구",
    4: "④ 영통구"
}

# 3. 응답 매핑 (1~5)
satisfaction_mapping = {
    1: "① 복지",
    2: "② 교통",
    3: "③ 산업 및 경제",
    4: "④ 기업 유치 및 일자리",
    5: "⑤ 주거",
    6: "⑥ 도시계획 및 정비",
    7: "⑦ 문화, 체육, 관광",
    8: "⑧ 안전",
    9: "⑨ 환경 및 기후변화",
    10: "⑩ 교육",
    11: "⑪ 아동․여성․가족",
    12: "⑫ 기타"
}

# 4. 구/만족도 매핑 적용
dm5_1['구'] = dm5_1['GU'].map(gu_mapping)
# q6(앞으로 수원시가 어느 정책 분야를 우선적으로 추진해야한다고 생각하나?)
dm5_1['만족도'] = dm5_1['q6'].map(satisfaction_mapping)

# 5. 교차표 만들기 (구 × 만족도)
cross_tab = pd.crosstab(dm5_1['구'], dm5_1['만족도']).reindex(index=gu_mapping.values(), columns=satisfaction_mapping.values(), fill_value=0)

# 6. 출력
#print(cross_tab)

# 비율(%)로 변환 후 소수점 2자리로 반올림
cross_tab_percent = cross_tab.div(cross_tab.sum(axis=1), axis=0).round(4) * 100

# 결과 출력
print(cross_tab_percent)

```

```python
# 1. DM5가 1인 사람만 필터링
dm5_1 = df_24_4[(df_24_4['DM5'] == 1) & (df_24_4['DM2'].between(19, 35))]


# 2. 응답 텍스트 매핑 딕셔너리
mapping = {
    1: "① 태어나서 자란 도시",
    2: "② 태어나지 않았지만 고향이나 다름없는 도시",
    3: "③ 마음을 붙이고 살만한 도시",
    4: "④ 기회가 있으면 떠나고 싶은 도시",
    5: "⑤ 잠시 머무르는 도시",
    6: "⑥ 기타",
}

# 3. q14(귀하에게 '수원'은 어떤 의미를 주는 도시?) 응답값별로 카운트
mapped = dm5_1['q14'].map(mapping)

# 4. 카운트 + 보기 순서대로 정렬
q9_counts = mapped.value_counts().reindex(mapping.values(), fill_value=0)

# 5. 출력
print(q9_counts)
```

```python
# 2. 응답 텍스트 매핑 딕셔너리
mapping = {
    1: "① 전혀 동의하지 않는다",
    2: "② 동의하지 않는다",
    3: "③ 보통이다",
    4: "④ 동의한다",
    5: "⑤ 매우 동의한다"
}


# 3. q17_5(귀하는 '수원'을 어떻게 평가함_교통이 편리한 도시?) 응답값별로 카운트
mapped = dm5_1['q17_5'].map(mapping)

# 4. 카운트 + 보기 순서대로 정렬
q9_counts = mapped.value_counts().reindex(mapping.values(), fill_value=0)

# 5. 출력
print(q9_counts)
```

```python
import pandas as pd

# .xlsx 파일 읽기
df = pd.read_excel('수원시 도시정책지표 공개 데이터.xlsx', engine='openpyxl')

# 데이터프레임 사용은 CSV랑 동일
print(df.head())
```

```python
# 2. '청년인구 비율'인 행만 선택
청년인구행 = df[df.iloc[:, 1] == '청년인구 비율']

# 3. 연도 범위 (2010~2024년)
years = list(range(2010, 2025))  # [2010, ..., 2024]

# 4. 연도 데이터만 따로 추출 (Unnamed: 5 ~ Unnamed: 19 → 열 인덱스 5~19)
청년인구연도값 = 청년인구행.iloc[0, 5:20].astype(float)

# 5. Series로 만들고, 연도 라벨 붙이기
청년인구_series = pd.Series(data=청년인구연도값.values, index=years, name="청년인구 비율")

# 6. 결과 출력
print(청년인구_series)


## 청년인구 비율 자체는 점점 줄어듬
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(청년인구_series.index, 청년인구_series.values, marker='o', linestyle='-', color='mediumblue')


plt.title("Youth Population Ratio in Suwon (2010–2024)")
plt.xlabel("Year")
plt.ylabel("Youth Population Ratio (%)")

plt.ylim(22, 27)


plt.xticks(청년인구_series.index, rotation=45)
plt.grid(True)


plt.tight_layout()
plt.show()
```
